```r
# Carregar o pacote necessário
library(readxl)

# Especifique o caminho do arquivo Excel
caminho_do_arquivo <- "base mortalidade.xlsx"

# Ler o arquivo Excel
dados <- read_excel(caminho_do_arquivo)

# Exibir os primeiros registros dos dados
print(head(dados))
```

```r
# Removendo as linhas que contêm "Ignorado anos" na coluna 'MORTE POR IDADE'
dados <- subset(dados, dados[['MORTE POR IDADE']] != "Ignorado anos")
```

```r
# Contando as categorias únicas na coluna 'MORTE POR IDADE'
contagem_categorias <- table(dados[['MORTE POR IDADE']])
print(contagem_categorias)

```

```r
categorias_unicas_ordenadas <- sort(unique(dados[['MORTE POR IDADE']]))
print(categorias_unicas_ordenadas)
```

```r
tamanho <- length(categorias_unicas_ordenadas)
print(tamanho)
```

```r
# Supondo que a coluna 'MORTE POR IDADE' é do tipo character e precisa ser convertida para numeric
dados[['MORTE POR IDADE']] <- as.numeric(gsub(" anos", "", dados[['MORTE POR IDADE']]))
```

```r
# Substituir NA por "sem informação" na coluna 'Escolaridade ajt'
dados$'Escolaridade ajt' <- ifelse(is.na(dados$'Escolaridade ajt'), "sem informação", dados$'Escolaridade ajt')
```

```r
# Calcular a amplitude
amplitude <- max(dados[['MORTE POR IDADE']]) - min(dados[['MORTE POR IDADE']])

# Calcular o número de classes pela Regra de Sturges
num_classes <- 6

# Calcular a amplitude dos intervalos de classe
amplitude_intervalo <- ceiling(amplitude / num_classes)

# Calcular os limites dos intervalos de classe
limites <- seq(min(dados[['MORTE POR IDADE']]), max(dados[['MORTE POR IDADE']]) + amplitude_intervalo, by = amplitude_intervalo)

# Criar uma nova coluna com as faixas de distribuição de frequência da idade
dados$Faixa_Idade <- cut(dados[['MORTE POR IDADE']], breaks = limites, include.lowest = TRUE, labels = paste0(limites[-length(limites)], " - ", limites[-1]))

# Visualizar as primeiras linhas do dataframe com a nova coluna
head(dados)
```

```r
# Supondo que 'Faixa_Idade' e 'Escolaridade ajt' estejam preenchidas em 'dados'

# Realizar o teste qui-quadrado de independência
resultado_chi2 <- chisq.test(dados[['Faixa_Idade']], dados[['Escolaridade ajt']])

# Visualizar o resultado do teste
print(resultado_chi2)
```

```r
# Fazer a contagem da coluna Escolaridade_ajt
contagem_doenca<- table(dados$'Doença ajt')

# Exibir a contagem
print(contagem_doenca)
```

## Split dos dados

```r
# Criar dataframe para Cirrose
cirrose_df <- dados[dados$`Doença ajt` == "Cirrose", ]

# Criar dataframe para Câncer
cancer_df <- dados[dados$`Doença ajt` == "Câncer", ]
```

## Testes de Hipotese só com o Cirrose

```r
# Realizar o teste qui-quadrado de independência
cirrose_resultado_chi2 <- chisq.test(cirrose_df[['Faixa_Idade']], cirrose_df[['Escolaridade ajt']])

# Visualizar o resultado do teste
print(cirrose_resultado_chi2)
```

```r
# Criar uma tabela de contingência
tabela_contingencia_cirrose <- table(cirrose_df$`Faixa_Idade`, cirrose_df$`Escolaridade ajt`)

# Visualizar a tabela de contingência
print(tabela_contingencia_cirrose)

# Realizar o Teste Exato de Fisher com simulação de Monte Carlo
fisher_cirrose <- fisher.test(tabela_contingencia_cirrose, simulate.p.value = TRUE, B = 10000)

# Visualizar os resultados do teste
print(fisher_cirrose)
```

```r
# Remover caracteres não numéricos da coluna MORTE POR IDADE
cirrose_df$`MORTE POR IDADE` <- as.numeric(gsub("[^0-9.]", "", cirrose_df$`MORTE POR IDADE`))

# Verificar se a conversão foi bem-sucedida
str(cirrose_df$`MORTE POR IDADE`)

# Criar o histograma novamente
hist(cirrose_df$`MORTE POR IDADE`,
     main = "Histograma de Mortes por Idade Cirrose",
     xlab = "Idade",
     ylab = "Frequência",
     col = "blue",
     border = "black")
```

```r
cirrose_morte_uf <- table(cirrose_df$`Municipio ajt`)

cor.test(cirrose_morte_uf, ibge$`IDH [2021]`, method = "pearson")
```

## Testes de Hipotese só com o Câncer

```r
# Realizar o teste qui-quadrado de independência
cancer_resultado_chi2 <- chisq.test(cancer_df[['Faixa_Idade']], cancer_df[['Escolaridade ajt']])

# Visualizar o resultado do teste
print(cancer_resultado_chi2)
```

```r
# Criar uma tabela de contingência
tabela_contingencia_cancer <- table(cancer_df$`Faixa_Idade`, cancer_df$`Escolaridade ajt`)

# Visualizar a tabela de contingência
print(tabela_contingencia_cancer)

# Realizar o Teste Exato de Fisher com simulação de Monte Carlo
fisher_cancer <- fisher.test(tabela_contingencia_cancer, simulate.p.value = TRUE, B = 10000)

# Visualizar os resultados do teste
print(fisher_cancer)
```

```r
# Remover caracteres não numéricos da coluna MORTE POR IDADE
cancer_df$`MORTE POR IDADE` <- as.numeric(gsub("[^0-9.]", "", cancer_df$`MORTE POR IDADE`))

# Verificar se a conversão foi bem-sucedida
str(cancer_df$`MORTE POR IDADE`)

# Criar o histograma novamente
hist(cancer_df$`MORTE POR IDADE`,
     main = "Histograma de Mortes por Idade Câncer",
     xlab = "Idade",
     ylab = "Frequência",
     col = "Red",
     border = "black")
```

```r
cancer_morte_uf <- table(cancer_df$`Municipio ajt`)

cor.test(cancer_morte_uf, ibge$`IDH [2021]`, method = "pearson")
```
